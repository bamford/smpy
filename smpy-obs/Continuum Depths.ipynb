```python
%pylab inline
import matplotlib.pyplot as plt
import numpy
from astropy.io import fits
from astropy.cosmology import Planck13
from mkCSPs import *

z = 4.1
Gal = CSP()
Gal.build(0.1,1,0,4,sfh_law='pow',fesc=0.2) # 100 Myr Constant SFR - Solar Metallicity, + Nebular continuum emission
Gal.addEmissionLine(1216,100) # Add Lyman-alpha Emission of desired Equivalent width 

# This section just loads in the SHARDS filters and makes 1500A tophat to calculate rest-frame UV mag
Shards = FilterSet('transmCurves_SHARDS/*.res')
wavelengths = numpy.array([filt.lambda_c for filt in Shards.filters])
UVfilt = FilterSet()
UVfilt.addTophatFilter(1500,150)

#Load Cluster Magnification map
hdulist = fits.open("hlsp_frontier_model_macs1149_cats_v1_z04-magnif.fits")
z4mag = hdulist[0].data
```

### First, i am normalising the galaxy such that the continuum at 1500 Angstroms (from closest SHARDS filter) is equal to the expected SHARDS depths:

```python
UVrest = Observe(Gal,UVfilt,0.)
S_z4 = Observe(Gal,Shards,z) 

print UVrest.AB
print S_z4.AB[15]

depth = 27. #3-sigma
fluxdepth = 10**((23.9 - depth)/2.5) * U.Jy

galflux = S_z4.fluxes[15] * 1e6

print fluxdepth/galflux

Gal *= fluxdepth/galflux
```

### 'Observe' the normalised galaxy to measure its rest-frame UV mag and observed SHARDS fluxes/magnitudes:

```python
UVrest = Observe(Gal,UVfilt,0.)
S_z4 = Observe(Gal,Shards,z) 

plt.semilogx([4000,10000],[27.,27.],'--',color='0.8')
plt.semilogx(wavelengths,S_z4.AB,'o')
plt.ylim([28.5,25.])
plt.xlim([4e3,1e4])
plt.xlabel(r'Wavelength, $\rm{\AA}$')
plt.ylabel('AB Magnitude')
```

### Plot the rest-frame UV limits after predicted cluster magnification:

```python
z4cut = z4mag[400:1100,400:1100]

z4cut = numpy.minimum(100,z4mag[400:1000,400:1000])

obslim = UVrest.AB.value

maglim = obslim + (2.5*numpy.log10(1+z4cut))


fig = plt.figure(figsize=(6,6))
Ax = fig.add_subplot(111)
Im = Ax.imshow(maglim,origin='lower',cmap=plt.cm.YlGnBu)
CB = fig.colorbar(Im,aspect=15,orientation='horizontal')
CB.set_label('Rest-frame UV Magnitude')
Ax.set_title('Rest-frame UV Magnitude Limit',size=14)
```
